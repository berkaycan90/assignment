```python


from sklearn.preprocessing import LabelEncoder


df = pd.read_csv("/Users/amir/Downloads/satis cs3.csv", dtype={'column0': str, 'column32': str}, low_memory=False)


df = df.dropna()


olist = ['column5', 'column6', 'column6']


lbl_encode = LabelEncoder()
for i in olist:
    df[i] = lbl_encode.fit_transform(df[i])
```

```python


plt.scatter(df["Primary Segment (Belge)"], df["All Parts USD	Total CS Revenue EUR"])

plt.xlabel("Primary Segment (Belge)")
plt.ylabel("All Parts USD	Total CS Revenue EUR")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
scaler = StandardScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)
```

```python
#az ChatGPT bepors
kmeans_model = KMeans(n_clusters = 4)
kmeans_model.fit(scaled_data)
```

```python
df["clusters"] = kmeans_model.labels_
df.head()
```

```python
Untitled.ipynb
plt.scatter(df["CompanyName"], 
            df["TotalPrice"], 
            c = df["clusters"])
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(scaled_data, kmeans_model.labels_, metric='euclidean')
```
